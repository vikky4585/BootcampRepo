```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Load in csv
df_rain = pd.read_csv('../Resources/avg_rain_state.csv')
df_rain.head()
```

```python
# Set x axis and tick locations
x_axis = np.arange(len(df_rain))

ticks = [(x + 0.4) for x in x_axis ]

y_axis = df_rain['Inches']
plt.figure(figsize=(20,3))
plt.bar(x_axis,y_axis,color='green', alpha=0.75,align='center')
```

```python
# Create a list indicating where to write x labels and set figure size to adjust for space
#plt.xlim(0,100)

plt.xticks(ticks,df_rain['State'], rotation='vertical')
```
