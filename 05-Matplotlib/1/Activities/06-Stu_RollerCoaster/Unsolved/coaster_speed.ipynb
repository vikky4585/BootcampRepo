```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
danger = [9, 8, 90, 85, 80, 70, 70, 65, 55, 60, 70, 65, 50]
railgun = [75, 70, 60, 65, 60, 45, 55, 50, 40, 40, 35, 35, 30]
time = np.arange(0,130,10)
time
```

```python
plt_danger = plt.plot(time,danger,color='Red',label='Danger')
plt_gun =  plt.plot(time,railgun,color='Green',label='RailGun')
plt.grid()
plt.xlabel('Coaster Runtime')
plt.ylabel('Speed MPH')
plt.title('Coaster Speed Over Time')
plt.ylim(5,95)
plt.xlim(0,120)
```
