```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy import stats
from scipy.stats import linregress
```

```python

df = pd.read_csv('../Resources/crime_data.csv')
years = df['year']
crime_rate = df['violent crime rate']

(slope,inter,_,_,_,)=linregress(years,crime_rate)
fit = slope * years + inter
fit
```

```python
fig,ax = plt.subplots()
ax.plot(years,crime_rate, marker='o')
ax.plot(years,fit, 'b--')
```
