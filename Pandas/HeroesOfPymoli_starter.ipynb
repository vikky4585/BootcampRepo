### Heroes Of Pymoli Data Analysis
* Of the 1163 active players, the vast majority are male (84%). There also exists, a smaller, but notable proportion of female players (14%).

* Our peak age demographic falls between 20-24 (44.8%) with secondary groups falling between 15-19 (18.60%) and 25-29 (13.4%).

```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# Raw data file
file_to_load = "Resources/purchase_data.csv"

# Read purchasing file and store into pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

## Player Count

* Display the total number of players


```python
unique_total_players = len(purchase_data['SN'].unique())
print(unique_total_players)
total_players = purchase_data['SN'].count()
print(total_players)
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
unique_items = purchase_data['Item ID'].nunique()
avg_price = purchase_data['Price'].mean()

df_summary_purchase = pd.DataFrame({"Unique Items":[unique_items],"Average Price" : [avg_price]})
df_summary_purchase['Average Price'] = df_summary_purchase['Average Price'].map("{:.2f}".format)
df_summary_purchase
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python

df_gender = purchase_data.groupby('Gender').SN.nunique().reset_index(name='counts')

#df_byGender = purchase_data.groupby('Gender').size().reset_index(name='counts')

female_count = df_gender.loc[(df_gender['Gender'] == 'Female'),:].reset_index().at[0,'counts']
female_percent = (female_count/unique_total_players) * 100
male_count = df_gender.loc[(df_gender['Gender'] == 'Male'),:].reset_index().at[0,'counts']
male_percent = (male_count/unique_total_players) * 100
other_count = df_gender.loc[(df_gender['Gender'] != 'Male') & (df_gender['Gender'] != 'Female') ,:].reset_index().at[0,'counts']
other_percent = (other_count/unique_total_players) * 100
print(f"Female Count {str(female_count)} Female Percentage {str(female_percent)}")
print(f"Male Count {str(male_count)} Male Percentage {str(male_percent)}")
print(f"Other Count {str(other_count)} Female Percentage {str(other_percent)}")

```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
df_bygender = purchase_data.groupby('Gender')
df_bygender.head()


df_bygender_summary = pd.merge(df_bygender.size().reset_index(name='counts'),
                               df_bygender['Price'].mean().reset_index(name='Avg Purchase Price'),
                              on='Gender')
df_bygender_summary['Avg Purchase Price'] = df_bygender_summary['Avg Purchase Price'].map("{:.2f}".format)
df_bygender_summary
#df_bygender
#df_gender_person = purchase_data.groupby(['Gender','SN'])
#df_gender_person.head()
# TODO: avg. purchase total per person --> what does this mean
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Establish bins for ages
age_bins = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99999]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
df_withagegroup = purchase_data.copy()
df_withagegroup['Age Group'] = pd.cut(purchase_data['Age'],age_bins,labels=group_names)
df_withagegroup = df_withagegroup.groupby('Age Group')


df_summary_age = pd.DataFrame({'Number': df_withagegroup['Age Group'].count() ,
                               "Percent":(df_withagegroup['Age Group'].count() / total_players) * 100})

df_summary_age['Percent'] = df_summary_age['Percent'].map("{:.2f}%".format)
df_summary_age
 
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
df_agegroup = purchase_data.copy()
df_agegroup['Age Group'] = pd.cut(purchase_data['Age'],age_bins,labels=group_names)
#df_agegroup = df_agegroup.groupby(['Age Group','SN'])
df_agegroup = df_agegroup.groupby('Age Group')
df_agegroup_summary = pd.merge(df_agegroup.size().reset_index(name='Purchase Count'),
                              df_agegroup['Price'].mean().reset_index(name='Avg Purchase Price'),
                              on='Age Group')
df_agegroup_summary['Avg Purchase Price'] = df_agegroup_summary['Avg Purchase Price'].map('{:.2f}'.format)
df_agegroup_summary
# TODO: avg. purchase total per person --> what does this mean
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
df_items = purchase_data[['Item ID','Item Name','Price']]
df_items = df_items.groupby(['Item ID','Item Name'])

df_items_summary = pd.DataFrame({'Purchase Count':df_items['Item ID'].count(),
    'Item Price': df_items['Price'].mean(),
    'Total Purchase': df_items['Price'].sum()                            
})

#df_items_summary = df_items_summary[['Item ID','Item Name','Purchase Count','Item Price','Total Purchase']]
df_items_summary_bycount = df_items_summary.sort_values('Purchase Count', ascending=False)
df_items_summary_bycount
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
df_items_summary_byvalue = df_items_summary.sort_values('Total Purchase', ascending=False)
df_items_summary_byvalue
```
