# PyCity Schools Analysis

* As a whole, schools with higher budgets, did not yield better test results. By contrast, schools with higher spending per student actually (\$645-675) underperformed compared to schools with smaller budgets (<\$585 per student).

* As a whole, smaller and medium sized schools dramatically out-performed large sized schools on passing math performances (89-91% passing vs 67%).

```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
math_pass = (student_data.loc[(student_data['math_score']) >= 70,['math_score']].count() * 100) / student_data['Student ID'].count()

reading_pass = (student_data.loc[(student_data['reading_score']) >= 70,['reading_score']].count() * 100) / student_data['Student ID'].count()

df_district_summary = pd.DataFrame({
    'Total Schools': [school_data['school_name'].count()],
    'Total Students': [student_data['Student ID'].count()],
    'Total Budget': pd.Series(school_data['budget'].sum()).map('{:,.2f}'.format),
    'Avg Math Score': pd.Series(student_data['math_score'].mean()).map('{:.2f}'.format),
    'Avg Reading Score': pd.Series(student_data['reading_score'].mean()).map('{:.2f}'.format),
    'Overall Avg Score': pd.Series((student_data['reading_score'].mean() + student_data['math_score'].mean())/2).map('{:.2f}'.format),
    'Math Pass %': pd.Series(math_pass.reset_index(name='pass').at[0,'pass']).map('{:.2f}%'.format),
    'Reading Pass %': pd.Series(reading_pass.reset_index(name='pass').at[0,'pass']).map('{:.2f}%'.format),
     
})
                        
df_district_summary
```

```python
df_school_level = school_data_complete.groupby('school_name')

df_math_pass = school_data_complete.loc[school_data_complete['math_score'] >= 70]
df_math_pass = df_math_pass.groupby('school_name')

df_reading_pass = school_data_complete.loc[school_data_complete['reading_score'] >= 70]
df_reading_pass = df_reading_pass.groupby('school_name')
#pass_byschool = df_school_level.filter(lambda x: (x['math_score'] >= 70).any())
#pass_byschool = df_school_level.filter(lambda x: x['math_score'] >= 70)
math_pass_byschool = (df_math_pass['math_score'].count() * 100) / df_school_level['Student ID'].count()
reading_pass_byschool = (df_reading_pass['reading_score'].count() * 100) / df_school_level['Student ID'].count()
```

## Top Performing Schools (By Passing Rate)

* Sort and display the top five schools in overall passing rate

```python

df_school_summary = pd.DataFrame({

    'School Type' : df_school_level['type'].first(),
    'Total Students':df_school_level['Student ID'].count(),
    'School Budget': pd.Series(df_school_level['budget'].first()).map('{:,.2f}'.format),
    'Per Student Budget': df_school_level['budget'].first()/df_school_level['Student ID'].count(),
    'Avg Math Score': pd.Series(df_school_level['math_score'].sum()/df_school_level['Student ID'].count()).map('{:.2f}'.format),
    'Avg Reading Score':pd.Series(df_school_level['reading_score'].sum()/df_school_level['Student ID'].count()).map('{:.2f}'.format),
    'Math Pass %': math_pass_byschool,
    'Reading Pass %':reading_pass_byschool,
    'Overall Passing Rate %':pd.Series((math_pass_byschool + reading_pass_byschool)/2).map('{:.2f}'.format),

})

df_school_summary['Math Pass %'] = df_school_summary['Math Pass %'].map('{:.2f}'.format)
df_school_summary['Reading Pass %'] = df_school_summary['Reading Pass %'].map('{:.2f}'.format)
df_school_summary
```

## Bottom Performing Schools (By Passing Rate)

* Sort and display the five worst-performing schools

```python
df_school_summary_sorted = df_school_summary.sort_values(by='Overall Passing Rate %')
df_school_summary_sorted.head()
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
df9  =  student_data.loc[student_data['grade'] == '9th' ].groupby('school_name')['math_score'].mean().map("{:.2f}".format)
df10  =  student_data.loc[student_data['grade'] == '10th' ].groupby('school_name')['math_score'].mean().map("{:.2f}".format)
df11  =  student_data.loc[student_data['grade'] == '11th' ].groupby('school_name')['math_score'].mean().map("{:.2f}".format)
df12  =  student_data.loc[student_data['grade'] == '12th' ].groupby('school_name')['math_score'].mean().map("{:.2f}".format)


df_math_avg = pd.concat([df9,df10,df11,df12],axis=1)
df_math_avg.columns=['9th Math Avg','10th Math Avg','11th Math Avg','12th Math Avg']
df_math_avg
```

## Reading Score by Grade 

* Perform the same operations as above for reading scores

```python
df9  =  student_data.loc[student_data['grade'] == '9th' ].groupby('school_name')['reading_score'].mean().map("{:.2f}".format)
df10  =  student_data.loc[student_data['grade'] == '10th' ].groupby('school_name')['reading_score'].mean().map("{:.2f}".format)
df11  =  student_data.loc[student_data['grade'] == '11th' ].groupby('school_name')['reading_score'].mean().map("{:.2f}".format)
df12  =  student_data.loc[student_data['grade'] == '12th' ].groupby('school_name')['reading_score'].mean().map("{:.2f}".format)


df_reading_avg = pd.concat([df9,df10,df11,df12],axis=1)
df_reading_avg.columns=['9th Reading Avg','10th Reading Avg','11th Reading Avg','12th Reading Avg']
df_reading_avg
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Sample bins. Feel free to create your own bins.
spending_bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]
df_school_spending = df_school_summary[['Avg Math Score','Avg Reading Score',
                                        'Math Pass %','Reading Pass %','Overall Passing Rate %'
                                        ,'Per Student Budget']]
df_school_spending['Spending Range'] = pd.cut(df_school_spending['Per Student Budget'], bins=spending_bins, labels=group_names)
df_school_spending = df_school_spending.reset_index()
df_school_spending['Avg Math Score'] = pd.to_numeric(df_school_spending['Avg Math Score'])
df_school_spending['Avg Reading Score'] = pd.to_numeric(df_school_spending['Avg Reading Score'])
df_school_spending['Math Pass %'] = pd.to_numeric(df_school_spending['Math Pass %'])
df_school_spending['Reading Pass %'] = pd.to_numeric(df_school_spending['Reading Pass %'])
df_school_spending['Overall Passing Rate %'] = pd.to_numeric(df_school_spending['Overall Passing Rate %'])

df_school_spending = df_school_spending.groupby('Spending Range')

df_spending_summary = pd.DataFrame({
    'Avg Math Score':df_school_spending['Avg Math Score'].mean(),
    'Avg Reading Score':df_school_spending['Avg Reading Score'].mean(),
    'Math Pass %':df_school_spending['Math Pass %'].mean(),
    'Reading Pass %':df_school_spending['Reading Pass %'].mean(),
    'Overall Passing Rate %':df_school_spending['Overall Passing Rate %'].mean()

}
)

df_spending_summary

```
