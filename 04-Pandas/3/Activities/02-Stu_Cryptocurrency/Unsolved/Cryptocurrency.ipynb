```python
# Import Dependencies
import pandas as pd
```

```python
bitcoin_csv = "Resources/bitcoin_cash_price.csv"
dash_csv = "Resources/dash_price.csv"
```

```python
bitcoin_df = pd.read_csv(bitcoin_csv)
dash_df = pd.read_csv(dash_csv)
```

```python
bitcoin_df.head()
```

```python
dash_df.head()
```

```python
# Merge the two DataFrames together based on the Dates they share
df_merge = pd.merge(bitcoin_df,dash_df, on="Date")
df_merge.head()
```

```python
# Rename columns so that they are differentiated
```

```python
# alternatively you can set your suffixes when the merge occurs
```

```python
# Collecting best open for Bitcoin and Dash
open_bitcoin = max(df_merge['Open_x'])
open_dash = max(df_merge['Open_y'])

# Collecting best close for Bitcoin and Dash

close_bitcoin = max(df_merge['Close_x'])
close_dash = max(df_merge['Close_y'])


# Collecting the total volume for Bitcoin and Dash
total_bitcoin = df_merge['Volume_x'].sum()
total_dash = df_merge['Volume_y'].sum()
```

```python
# Creating a summary DataFrame using above values

pd.DataFrame(
    {'Best Open Bitcoin': [open_bitcoin], 'Best Open Dash': [open_dash],'Best Close BT':[close_bitcoin],
             'Best Close Dash' : [close_dash],'Total Volume BT': [total_bitcoin],'Total Volume Dash': [total_dash]}
)
```
