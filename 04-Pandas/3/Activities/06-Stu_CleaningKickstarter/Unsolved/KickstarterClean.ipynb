```python
import pandas as pd
```

```python
# The path to our CSV file

# Read our Kickstarter data into pandas
df = pd.read_csv("Resources/KickstarterData.csv")
df.head()
```

```python
# Get a list of all of our columns for easy reference
print(df.columns)
```

```python
# Extract "name", "goal", "pledged", "state", "country", "staff_pick",
# "backers_count", and "spotlight"
df_extract = df[["name","goal","pledged", "state", "country", "staff_pick","backers_count","spotlight"]]
# "]]
df_extract.head()
df_extract.count()
```

```python
# Remove projects that made no money at all
df_havemoney = df_extract.loc[df_extract['pledged'] > 0,:]
df_havemoney.count()
```

```python
# Collect only those projects that were hosted in the US
df_US = df_havemoney.loc[df_havemoney['country'] == 'US',:]
df_US.count()
```

```python
# Create a new column that finds the average amount pledged to a project
df_US['average_pledge'] = df_US['pledged'] / df_US['backers_count']
df_US.head()
```

```python
# First convert "average_donation", "goal", and "pledged" columns to float
#pd.to_numeric(df_US['average_pledge'])
#pd.to_numeric(df_US['goal'])
#pd.to_numeric(df_US['pledged'])

# Then Format to go to two decimal places, include a dollar sign, and use comma notation
#df_US['average_pledge'] = df_US['average_pledge'].map("${:.2f}".format)
df_US['average_pledge'] = df_US['average_pledge'].astype(float).map("${:,.2f}".format)
df_US['goal'] = df_US['goal'].astype(float).map("${:,.2f}".format)
df_US['pledged'] = df_US['pledged'].astype(float).map("${:,.2f}".format)
df_US
```

```python
# Calculate the total number of backers for all US projects
df_US['backers_count'].sum()
```

```python
# Calculate the average number of backers for all US projects
df_US['backers_count'].mean()
```

```python
# Collect only those US campaigns that have been picked as a "Staff Pick"
df_staff = df_US.loc[df_US['staff_pick'] == True,:]
df_staff.head()
```

```python
# Group by the state of the campaigns and see if staff picks matter (Seems to matter quite a bit)
df_grouped = df_staff.groupby('state')
df_grouped['name'].count()
```
