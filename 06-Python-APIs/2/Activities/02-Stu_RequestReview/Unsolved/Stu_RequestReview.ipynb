```python
# Dependencies
import json
import requests 
```

```python
# Specify the URL
url = 'http://nyt-mongo-scraper.herokuapp.com/api/headlines'

# Make request and store response
response = requests.get(url)
```

```python
# JSON-ify response
response = response.json()
response
```

```python
# Print first and last articles
first = response[0]['headline']
print(first)
last = response[len(response) -1]['headline']
last = response[-1]['headline']

print(last)
```
